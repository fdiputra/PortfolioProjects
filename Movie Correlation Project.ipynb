```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the configuration of the plot we will create

# Read in the data

df = pd.read_csv(r'C:\Users\admin\Downloads\movies.csv')
df=df.dropna()
```

```python
# Lets look at the data 
df.head()
```

```python
# Lets see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(pct_missing*100)))
```

```python
# Data types for our columns
df.dtypes
```

```python
#Change Data type of columns
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.head
```

```python
#Create correct year column
df['yearcorrect']=df['released'].str.extract(pat='([0-9]{4})').astype(int)
```

```python
df=df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#Drop any duplicate
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Gross Earnings')
plt.ylabel('Bugdet for Film')
plt.show()
```

```python
#Plot budget vs gross using seaborn

sns.regplot(x='budget',y='gross',data=df,scatter_kws= {"color":"red"},line_kws={"color":"blue"})
```

```python
#Let's start looking at correlation

df.corr(method='pearson') #pearson.kendall,spearman
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot= True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Looks at company
df_numerized= df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head
        
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot= True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs 
```

```python
high_corr = sorted_pairs[(sorted_pairs) >0.5]
high_corr
```
